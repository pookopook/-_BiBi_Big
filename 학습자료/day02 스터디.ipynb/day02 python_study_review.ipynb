### 컬렉션 데이터 타입
 - list: 순서가 있는 요소의 컬렉션을 저장, 요소는 대괄호 [] 안에 쉼표로 구분되어 나열됩니다

 - tuple: 순서가 있는 요소의 컬렉션을 저장하지만, 리스트와 달리 변경할 수 없습니다(immutable)

     - 세트는 중괄호 {}로 둘러싸이며, 요소는 쉼표로 구분됩니다 (예: {1, 2, 3, 3}, {'apple', 'banana', 'banana'})

# 리스트 함수의 종류
### 추가와 확장
 - append(x): 리스트의 끝에 항목 x를 추가합니다.
 - extend(iterable): 리스트의 끝에 주어진 iterable의 모든 항목을 추가합니다.
 - insert(i, x): 주어진 위치 i에 항목 x를 삽입합니다.
### 제거
 - remove(x): 리스트에서 첫 번째로 나타나는 항목 x를 제거합니다. 항목이 리스트에 없으면 오류를 발생시킵니다.
 - pop([i]): 주어진 위치 i에 있는 항목을 리스트에서 제거하고, 그 항목을 반환합니다. i가 지정되지 않으면, 리스트의 마지막 항목을 제거하고 반환합니다.
 - clear(): 리스트의 모든 항목을 제거합니다.
### 조회와 정렬
 - index(x[, start[, end]]): 항목 x가 나타나는 첫 번째 위치의 인덱스를 반환합니다. 선택적으로 start와 end를 지정하여 검색 범위를 제한할 수 있습니다.
 - count(x): 리스트에서 항목 x가 몇 번 나타나는지를 반환합니다.
 - sort(key=None, reverse=False): 리스트의 항목들을 정렬합니다. key 매개변수는 정렬 기준을 지정하는 데 사용되며, reverse=True는 내림차순 정렬을 의미합니다.
 - reverse(): 리스트의 요소 순서를 뒤집습니다.
### 복사
 - copy(): 리스트를 복사합니다.
 - deepcopy(): 리스트를 깊게 복사합니다.(중첩된 리스트의 경우 사용)

# 리스트 연습

```python
# 리스트 지정
fruits = ['apple', 'banana', 'cherry']
```

```python
# 리스트의 끝에 요소 추가
fruits.append('date')
# 이제 fruits는 ['apple', 'banana', 'cherry', 'date']
```

```python
# 다른 리스트의 요소를 현재 리스트에 추가
more_fruits = ['fig', 'grape']
fruits.extend(more_fruits)
# extend와 append의 차이를 알아보자.
# 이제 fruits는 ['apple', 'banana', 'cherry', 'date', 'fig', 'grape']
```

```python
# 지정된 인덱스에 요소 삽입
fruits.insert(1, 'kiwi')
# 이제 fruits는 ['apple', 'kiwi', 'banana', 'cherry', 'date', 'fig', 'grape']
```

```python
# 리스트에서 첫 번째로 나타나는 특정 값을 제거
fruits.remove('date')
# 이제 fruits는 ['apple', 'kiwi', 'banana', 'cherry', 'fig', 'grape']
```

```python
# 지정한 인덱스의 요소를 제거하고, 그 요소 반환
popped_fruit = fruits.pop(2)
# popped_fruit는 'banana'
# 이제 fruits는 ['apple', 'kiwi', 'cherry', 'fig', 'grape']
```

```python
# 리스트에서 특정 값의 첫 번째 인덱스 찾기
index_of_cherry = fruits.index('cherry')
# index_of_cherry는 2
```

```python
# 리스트에서 특정 값의 요소 수 세기
count_of_apple = fruits.count('apple')
# count_of_apple은 1
```

```python
# 리스트의 요소를 정렬
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
# numbers는 이제 [1, 1, 2, 3, 4, 5, 9]
```

```python
# 리스트의 요소 순서를 반대로 뒤집기
fruits.reverse()
# fruits는 이제 ['grape', 'fig', 'cherry', 'kiwi', 'apple']
```

```python
# 리스트의 복사본을 새로운 리스트로 반환
fruits_copy = fruits.copy()
# fruits_copy는 ['grape', 'fig', 'cherry', 'kiwi', 'apple']
```

### 파이썬 변수는... 그냥 주소를 나타낼 뿐이다.

```python
from copy import copy
from copy import deepcopy
original_list = [[1, 2, 3], [4, 5, 6]]
copy_list = copy(original_list)
deep_list = deepcopy(original_list)

# 원본 리스트의 내부 리스트를 변경해도 깊은 복사본에는 영향을 주지 않습니다.
original_list[0][0] = 'X'  # 리스트의 첫번째 항목의, 첫번째 항목을 X로 변경했을 때 일어나는 일
print("원래 리스트"     ,original_list)
print("카피된 리스트"   ,copy_list)
print("딥카피된 리스트" ,deep_list)  # [[1, 2, 3], [4, 5, 6]] 출력
```

원본 객체와 복사본이 내부 객체를 공유하기 때문에, 한 객체 내의 내부 객체를 수정하면 다른 객체에도 영향을 줄 수 있습니다.
복잡한 객체(예: 리스트의 리스트, 클래스 인스턴스 등)에 대해 완전한 복제본을 만들고자 할 때 사용합니다.

```python
# 좀 더 정확한 이해를 위해서.
list_A = [1, 2, 3]
list_B = [4, 5, 6]
original_list = [list_A, list_B]
copy_list = copy(original_list)
deep_list = deepcopy(original_list)

# 원본 리스트의 내부 리스트를 변경해도 깊은 복사본에는 영향을 주지 않습니다.
original_list[0][0] = 'X'  # 리스트의 첫번째 항목의, 첫번째 항목을 X로 변경했을 때 일어나는 일
print("원래 리스트"     ,original_list)
print("원래 리스트 A"     ,list_A)
print("원래 리스트 B"     ,list_B)
print("카피된 리스트"   ,copy_list)
print("딥카피된 리스트" ,deep_list)  # [[1, 2, 3], [4, 5, 6]] 출력
```
