```python
from faker import Faker
import pandas as pd
import random
from datetime import timedelta

fake = Faker('ko_KR')

# Sample data parameters
num_purchase_data = 100
num_items = 10
num_users = 20
num_shipments = 50

# Generate unique IDs
user_ids = [fake.uuid4() for _ in range(num_users)]
shipment_ids = [fake.uuid4() for _ in range(num_shipments)]
items = [fake.word() for _ in range(num_items)]
# Define Korean item names
korean_items = ['사과', '바나나', '오렌지', '포도', '수박', '배', '딸기', '체리', '복숭아', '자두']

# Regenerate purchase data with Korean items
purchase_data_korean = []
for _ in range(num_purchase_data):
    purchase_date = fake.date()
    purchase_data_korean.append({
        '일자': purchase_date,
        '품목': random.choice(korean_items),
        '수량': random.randint(1, 1000),
        '구매자id': random.choice(user_ids),
        '운송장id': random.choice(shipment_ids)
    })

# Generate user data
user_data = []
for user_id in user_ids:
    user_data.append({
        '구매자id': user_id,
        '이름': fake.name(),
        '나이': random.randint(18, 70),
        '성별': random.choice(['남성', '여성']),
        '주소': fake.address()
    })

# Generate shipment data
shipment_data = []
for shipment_id in shipment_ids:
    shipment_date = fake.date()
    purchase_date = pd.to_datetime(shipment_date) - timedelta(days=random.randint(0, 14))
    shipment_data.append({
        '운송장id': shipment_id,
        '현재위치': fake.city(),
        '배송기사': fake.name(),
        '예상도착일자': purchase_date.strftime('%Y-%m-%d')
    })

# Convert to DataFrames
purchase_df = pd.DataFrame(purchase_data_korean)
user_df = pd.DataFrame(user_data)
shipment_df = pd.DataFrame(shipment_data)

# Save to CSV files
purchase_df.to_csv('./data/purchase_data_korean.csv', index=False, encoding='cp949')
user_df.to_csv('./data/user_data.csv', index=False, encoding='cp949')
shipment_df.to_csv('./data/shipment_data.csv', index=False, encoding='cp949')
```
