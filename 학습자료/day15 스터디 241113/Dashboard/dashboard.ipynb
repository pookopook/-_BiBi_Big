```python
import pandas as pd
import plotly.express as px
from jinja2 import Template

# 샘플 데이터 생성
data1 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [10, 15, 20]})
data2 = pd.DataFrame({'Category': ['D', 'E', 'F'], 'Values': [5, 25, 15]})

# Plotly 차트 생성
fig1 = px.bar(data1, x='Category', y='Values', title="탭 1 - 샘플 데이터 1")
fig2 = px.bar(data2, x='Category', y='Values', title="탭 2 - 샘플 데이터 2")

# HTML로 변환된 차트
chart_html1 = fig1.to_html(full_html=False, include_plotlyjs='cdn')
chart_html2 = fig2.to_html(full_html=False, include_plotlyjs=False)  # 첫 차트에 포함되므로 여기서는 제외

# HTML 템플릿 정의
html_template = """
<!DOCTYPE html>
<html>
<head>
    <title>Plotly 대시보드</title>
    <style>
        {% raw %}
        /* 탭 스타일 */
        .tab {
            display: none;
        }
        .tab-button {
            display: inline-block;
            padding: 10px;
            cursor: pointer;
            background-color: #f1f1f1;
            margin-right: 5px;
            border: 1px solid #ccc;
        }
        .tab-button.active {
            background-color: #ddd;
        }
        {% endraw %}
    </style>
</head>
<body>
    <h1>대시보드</h1>
    
    <!-- 탭 버튼 -->
    <div>
        <div class="tab-button active" onclick="openTab(0)">탭 1</div>
        <div class="tab-button" onclick="openTab(1)">탭 2</div>
    </div>

    <!-- 탭 내용 -->
    <div id="tabs">
        <div class="tab" style="display: block;">{{ chart_html1 | safe }}</div>
        <div class="tab">{{ chart_html2 | safe }}</div>
    </div>

    <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
    <script>
        {% raw %}
        function openTab(tabIndex) {
            var tabs = document.getElementsByClassName("tab");
            var buttons = document.getElementsByClassName("tab-button");

            for (var i = 0; i < tabs.length; i++) {
                tabs[i].style.display = (i === tabIndex) ? "block" : "none";
                buttons[i].classList.toggle("active", i === tabIndex);
            }
        }
        {% endraw %}
    </script>
</body>
</html>
"""

# Jinja2 템플릿을 사용해 HTML 콘텐츠 생성
template = Template(html_template)
html_content = template.render(chart_html1=chart_html1, chart_html2=chart_html2)

# HTML 파일로 저장
with open("plotly_dashboard.html", "w") as file:
    file.write(html_content)

print("plotly_dashboard.html 파일이 생성되었습니다.")
```

```python
import pandas as pd
import plotly.express as px
from jinja2 import Template

# 샘플 데이터 생성
# 탭 1: 산점도 데이터
data_scatter = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 13, 17, 19]
})

# 탭 2: 그룹화된 막대 그래프 데이터
data_grouped_bar = pd.DataFrame({
    'Group': ['G1', 'G1', 'G1', 'G2', 'G2', 'G2'],
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Values': [10, 15, 20, 12, 18, 25]
})

# 탭 3: 다중 막대 그래프 데이터
data_multi_bar = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 20, 25],
    'Value2': [12, 18, 22, 28],
    'Value3': [14, 20, 24, 30]
})

# 탭 4: 선형 회귀 산점도 데이터
data_regression = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 4, 5, 4, 5]
})

# Plotly 차트 생성
# 탭 1: 산점도
fig_scatter = px.scatter(
    data_scatter, x='X', y='Y', color='Category',
    title="산점도 차트", labels={'X': 'X 값', 'Y': 'Y 값'}
)

# 탭 2: 그룹화된 막대 그래프
fig_grouped_bar = px.bar(
    data_grouped_bar, x='Category', y='Values', color='Group',
    barmode='group', title="그룹화된 막대 그래프"
)

# 탭 3: 다중 막대 그래프
fig_multi_bar = px.bar(
    data_multi_bar, x='Category', y=['Value1', 'Value2', 'Value3'],
    title="다중 막대 그래프"
)

# 탭 4: 선형 회귀 산점도
fig_regression = px.scatter(
    data_regression, x='X', y='Y', trendline='ols',
    title="선형 회귀 산점도"
)

# HTML로 변환된 차트
chart_html1 = fig_scatter.to_html(full_html=False, include_plotlyjs='cdn')
chart_html2 = fig_grouped_bar.to_html(full_html=False, include_plotlyjs=False)
chart_html3 = fig_multi_bar.to_html(full_html=False, include_plotlyjs=False)
chart_html4 = fig_regression.to_html(full_html=False, include_plotlyjs=False)

# 데이터 테이블 생성 (HTML)
table_html1 = data_scatter.to_html(index=False)
table_html2 = data_grouped_bar.to_html(index=False)
table_html3 = data_multi_bar.to_html(index=False)
table_html4 = data_regression.to_html(index=False)

# HTML 템플릿 정의
html_template = """
<!DOCTYPE html>
<html>
<head>
    <title>Plotly 대시보드</title>
    <style>
        {% raw %}
        /* 탭 스타일 */
        .tab-content {
            display: none;
        }
        .tab-button {
            display: inline-block;
            padding: 10px 20px;
            cursor: pointer;
            background-color: #f1f1f1;
            margin-right: 5px;
            border: 1px solid #ccc;
            border-bottom: none;
        }
        .tab-button.active {
            background-color: #ddd;
            font-weight: bold;
        }
        /* 차트와 테이블 레이아웃 */
        .chart-table-container {
            display: flex;
            flex-direction: row;
        }
        .chart-container {
            flex: 1;
            padding: 10px;
        }
        .table-container {
            flex: 1;
            padding: 10px;
            overflow: auto;
            max-height: 500px;
        }
        /* 테이블 스타일 */
        table {
            border-collapse: collapse;
            width: 100%;
        }
        th, td {
            border: 1px solid #ccc;
            text-align: center;
            padding: 8px;
        }
        th {
            background-color: #eee;
        }
        /* 탭 컨테이너 스타일 */
        #tabs {
            border: 1px solid #ccc;
            padding-top: 10px;
        }
        {% endraw %}
    </style>
</head>
<body>
    <h1>대시보드</h1>
    
    <!-- 탭 버튼 -->
    <div>
        <div class="tab-button active" onclick="openTab(0)">산점도</div>
        <div class="tab-button" onclick="openTab(1)">그룹화된 막대 그래프</div>
        <div class="tab-button" onclick="openTab(2)">다중 막대 그래프</div>
        <div class="tab-button" onclick="openTab(3)">선형 회귀 산점도</div>
    </div>

    <!-- 탭 내용 -->
    <div id="tabs">
        <div class="tab-content" style="display: block;">
            <div class="chart-table-container">
                <div class="chart-container">
                    {{ chart_html1 | safe }}
                </div>
                <div class="table-container">
                    <h3>데이터 테이블</h3>
                    {{ table_html1 | safe }}
                </div>
            </div>
        </div>
        <div class="tab-content">
            <div class="chart-table-container">
                <div class="chart-container">
                    {{ chart_html2 | safe }}
                </div>
                <div class="table-container">
                    <h3>데이터 테이블</h3>
                    {{ table_html2 | safe }}
                </div>
            </div>
        </div>
        <div class="tab-content">
            <div class="chart-table-container">
                <div class="chart-container">
                    {{ chart_html3 | safe }}
                </div>
                <div class="table-container">
                    <h3>데이터 테이블</h3>
                    {{ table_html3 | safe }}
                </div>
            </div>
        </div>
        <div class="tab-content">
            <div class="chart-table-container">
                <div class="chart-container">
                    {{ chart_html4 | safe }}
                </div>
                <div class="table-container">
                    <h3>데이터 테이블</h3>
                    {{ table_html4 | safe }}
                </div>
            </div>
        </div>
    </div>

    <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
    <script>
        {% raw %}
        function openTab(tabIndex) {
            var tabs = document.getElementsByClassName("tab-content");
            var buttons = document.getElementsByClassName("tab-button");

            for (var i = 0; i < tabs.length; i++) {
                tabs[i].style.display = (i === tabIndex) ? "block" : "none";
                buttons[i].classList.toggle("active", i === tabIndex);
            }
        }
        {% endraw %}
    </script>
</body>
</html>
"""

# Jinja2 템플릿을 사용해 HTML 콘텐츠 생성
template = Template(html_template)
html_content = template.render(
    chart_html1=chart_html1,
    chart_html2=chart_html2,
    chart_html3=chart_html3,
    chart_html4=chart_html4,
    table_html1=table_html1,
    table_html2=table_html2,
    table_html3=table_html3,
    table_html4=table_html4
)

# HTML 파일로 저장
with open("plotly_dashboard_2.html", "w", encoding='utf-8') as file:
    file.write(html_content)

print("plotly_dashboard.html 파일이 생성되었습니다.")
```
