# 네트워크?

네트워크 분석은 복잡한 시스템에서 요소 간의 관계와 상호작용을 이해하고 분석하는 방법입니다.
이러한 시스템은 **노드(Node)**와 *엣지(Edge)*로 구성된 *그래프(Graph)*로 표현됩니다.

- 노드(Node): 개별 객체나 사람 등을 나타냅니다.
- 엣지(Edge): 노드 간의 연결이나 관계를 나타냅니다.

# 네트워크 그래프 그리기 연습

```python
# 빈 그래프 생성
G = nx.Graph()

plt.figure(figsize=(4,2))   # 너무 크기가 크므로 줄여보자
plt.title('Network Graph 연습')
nx.draw(G)
```

- nx.Graph() 는 방향성이 없는 그래프를 생성해주는 명령어이다.  
- 방향성이 있는 그래프 G = nx.DiGraph()  
- 다중 엣지 무방향 그래프 G = nx.MultiGraph()  
- 다중 엣지   방향 그래프 G = nx.MultiDiGraph()

```python
plt.figure(figsize=(8, 2.5))
plt.title('network analysis')
nx.draw(G, with_labels=True, font_family=font_name)
#pos = nx.kamada_kawai_layout(G,scale=100)
#pos는 그래프의 레이아웃을 정해준다. nx.draw(G, pos, **kargs)
```

```python
# 그래프 함수를 만들어보자. *args, **kwargs 키워드 매개변수
def 그래프(그래프, **변수들):
    if 'with_labels' not in 변수들:
        변수들['with_labels'] = True
    plt.figure(figsize=(8, 4))
    plt.title('network analysisㄱ')
    nx.draw(그래프, **변수들, font_family=font_name)
    #return plt.show()
```

## 에고 그래프(ego)
ego란 나를 중심으로 한 그래프를 표현하는 방법이다.

```python
# 그래프 함수를 좀 수정해봅시다. pos가 필요합니다.
def 그래프(그래프, *변수, **변수들):
    if len(변수) > 0:  # <- 1추가될 부분
        pos = 변수[0]   # <- 2추가될 부분
    else:
        pos = nx.spring_layout(그래프)  # pos가 주어지지 않았을 때 기본 레이아웃 사용
    if 'with_labels' not in 변수들:
        변수들['with_labels'] = True
    plt.figure(figsize=(8, 4))
    plt.title('network analysisㄱ')
    nx.draw(그래프, pos, **변수들, font_family=font_name)
    #return plt.show()
```

```python
egoG = nx.ego_graph(G,3,1)
pos = nx.spring_layout(egoG)  # 레이아웃 지정
#pos = nx.kamada_kawai_layout(egoG)
그래프(egoG, pos)
nx.draw_networkx_nodes(egoG, pos, nodelist=[3], node_color='red')
plt.show()
```

### 모듈러리티
 - 네트워크의 커뮤니티 구조를 측정하는 지표
 - 의미: 노드들이 얼마나 뚜렷한 군집을 형성하는지 평가  



```python
# 모듈러리티?
from networkx.algorithms import community
# 커뮤니티 탐지 (Greedy Modularity 알고리즘 사용)
communities = community.greedy_modularity_communities(G)

# 커뮤니티를 리스트 형식으로 변환
community_list = [list(c) for c in communities]

# 모듈러리티 계산
modularity = community.modularity(G, communities)

# 결과 출력
print(f"모듈러리티: {modularity}")
print("커뮤니티 구성:")
for idx, comm in enumerate(community_list):
    print(f"커뮤니티 {idx + 1}: {comm}")
```

```python
group1 = community_list[0]
group2 = community_list[1]
```

```python
G.nodes()
```

```python
node_colors = []
i = 0
for node in G.nodes():
    if node in group1:
        node_colors.append('blue')
    elif node in group2:
        node_colors.append('red')
    else:
        node_colors.append('gray')
node_colors, len(node_colors)
```
