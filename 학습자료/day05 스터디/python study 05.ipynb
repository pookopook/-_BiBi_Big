## 데이터 스케일링의 종류
| 알고리즘                     | 설명                                                                                       | 장점                                                                                       | 단점                                                       |
|-----------------------------|------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------|------------------------------------------------------------|
| 표준화 (Standardization)    | 평균을 0, 표준 편차를 1로 맞춤. 각 특징을 정규 분포로 변환.                                   | 대부분의 모델에 유용하며, 특히 정규 분포를 따르는 데이터에 적합.                                 | 이상치에 민감할 수 있음.                                     |
| 최소-최대 스케일링 (Min-Max Scaling) | 데이터 값을 0과 1 사이로 변환. 최소값을 0, 최대값을 1로 변환.                                 | 모든 데이터가 동일한 범위로 변환되어, 이상치가 없는 경우 유용.                                    | 이상치의 영향을 많이 받을 수 있음.                                 |
| 로버스트 스케일링 (Robust Scaling)  | 중앙값을 0, IQR(interquartile range)을 1로 맞춤.                                        | 이상치의 영향을 줄이는 데 유용.                                                           | 데이터의 분포가 균일하지 않은 경우 비효율적일 수 있음.                     |
| 정규화 (Normalization)      | 데이터의 범위를 [0, 1] 또는 [-1, 1]로 조정. 각 특징의 최소값과 최대값을 기준으로 변환.                    | 간단하고 직관적이며, L1 또는 L2 정규화와 함께 사용 가능.                                      | 이상치의 영향을 많이 받을 수 있음.                                 |
| MaxAbs 스케일링              | 각 특징의 최대 절대값을 1로 변환. 양수와 음수가 모두 포함된 데이터를 스케일링할 때 유용.                    | 데이터의 중심을 유지하며, 양수와 음수가 포함된 데이터에 유용.                                 | 다른 스케일링 기법보다 덜 일반적이며, 특정 데이터셋에만 적합할 수 있음.           |

# 데이터셋 준비

```python
## 라이브러리 임포트
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
## 데이터셋 불러오기
data = load_diabetes()
x = data.data
y = data.target
x_nm = data.feature_names
df = pd.DataFrame(data=x, columns=x_nm)     
df['target'] = y

print("Diabetes Dataset:", df.shape)
#print(data.DESCR), data.info(), data.describe()
#df
```

```python
# 분할하기
x = df.iloc[:, :10]
y = df['target']

#데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x
                                                   ,y
                                                   ,test_size=0.3
                                                   #,stratify=y #stratify는 왜 필요없을까?
                                                   ,random_state=42)
# 랜덤 인덱스를 초기화
x_train = x_train.reset_index(drop=True) 
x_test  = x_test .reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test  = y_test .reset_index(drop=True)
# 잘 됐는지 확인
#x_test.shape, x_train.shape, y_test.shape, y_train.shape
```

# 회귀모형 LinearRegression

##### 간단한 회귀모형

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# model 1: LinearRegression
m1 = LinearRegression()

# x_train, x_test, y_train, y_test
m1.fit(x_train, y_train)

# 예측 및 평가
y_pred = m1.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE: {mse}, R2: {r2}')
```

```python
# model 2: Decision Tree
from sklearn.tree import DecisionTreeRegressor

m2 = DecisionTreeRegressor()
m2.fit(x_train, y_train)

#예측 및 평가
y_pred = m2.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE: {mse}, R2: {r2}')
```

- 결과해석

다중공선성이 높은 변수들을 제외해보고 (10 이상)

적합한 변수만을 선별하여 다시 회귀모형을 돌려보자.

- 단계적 변수 선택(Stepwise Variable Selection):

전진 선택법(Forward Selection), 후진 제거법(Backward Elimination), 단계적 선택법(Stepwise Selection)과 같은 기법들을 사용하여 변수를 선택하거나 제거합니다. 이 방법은 변수를 하나씩 추가하거나 제거하여 최종 모델을 만들기 위해 모델의 성능을 평가합니다.

```python
##### 코드샘플
import statsmodels.api as sm

def backward_elimination(X, y):
    num_features = X.shape[1]
    for i in range(num_features):
        model = sm.OLS(y, X).fit()
        p_values = model.pvalues
        max_p_value = p_values.max()
        if max_p_value > 0.05:  # 임계값을 설정하여 변수를 제거하는 조건을 정의합니다.
            max_p_idx = p_values.idxmax()
            X = X.drop(max_p_idx, axis=1)
        else:
            break
    return X

# X는 독립 변수들의 데이터프레임, y는 종속 변수의 데이터프레임 또는 배열입니다.
# 후진 제거법을 적용하여 변수를 선택하고 최종적으로 선택된 변수들을 반환합니다.
selected_features = backward_elimination(X, y)
```
