## 빅분기 실기준비
### 커닝페이퍼 만들기
 - 빅분기 실기, 시험시간 정말 깁니다.
 - 빅분기 실기를 완벽하게 준비한다는 것은 비전문가 입장에서 쉽지 않은 일입니다.
 - 전문가가 아닌 이상, 실제 시험시간은 두가지에 시간을 많이 씁니다.
   - 코드 디버그
   - 몰라서 멍때리기(!)
 - 커닝페이퍼를 만드는 방법을 알게 되면,
 - 몰라서 멍때리는 시간을, 도움말 찾기라도 할 수 있는 시간으로 바꿀 수 있습니다.

````python
# 코드 실행 시 제공 패키지 리스트 확인 가능

import pkg_resources 
import pandas 

OutputDataSet = pandas.DataFrame(sorted([(i.key, i.version) for i in pkg_resources.working_set])) 

print(OutputDataSet)
````

- 보통은 pandas, numpy, scikit-learn, scipy, statsmodels 씁니다.
- pandas, numpy는 실기준비과정에서 대부분 마스터하게 되구요,
- 도움말 기능은 주로 scikit-learn, scipy, statsmodels 활용에 필요할겁니다.

````python

import statsmodels
print(dir(statsmodels))   #print(dir(stats))

import statsmodels.formula
# from scipy import stats # scipy는 보통 모듈별로 호출한다.
# dir - 해당 객체가 가지고 있는 모든 변수, 메소드를 출력
print(dir(statsmodels.formula.api))   #print(dir(stats))
````

````python
#로지스틱 회귀 검색하는 법.
print(dir(statsmodels.formula.api.logit))
````

````python
# help
print(help(statsmodels.formula.api.logit))
````

````python
# __all__ 과 dir의 차이점
print(statsmodels.formula.api.__all__)
print(dir(statsmodels.formula.api))
````

````python
# __doc__
print(statsmodels.formula.api.logit.__doc__)
````

### 결론
 - 라이브러리를 찾는것은 dir, __all__ 활용하자
    ``` python
    import statsmodels
    print(dir(statsmodels))   # dir

    print(statsmodels.formula.api.__all__)   #__all__  (언더바 두개로 감싼다. __)
    
    ```

 - 함수는 help, __doc__로 검색하여 사용방법을 확인하자.

   ``` python
   print(statsmodels.formula.api.logit.__doc__)  # __doc__   (언더바 두개로 감싼다. __)
   print(help(statsmodels.formula.api.logit))    # help
   ```
